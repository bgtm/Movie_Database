On créer une fonction pour nettoyer la colonne overview et harmoniser la façon d'écrire les mots

```python
def clean_liste (c):
  c = c.str.replace(",","")
  c = c.str.replace("'"," ")
  c = c.str.replace('"'," ")
  c = c.str.replace("(","")
  c = c.str.replace(")","")
  c = c.str.replace(".","")
  c = c.str.replace(";","")
  c = c.str.replace(":","")
  c = c.str.upper()

  return c 

l_descp = clean_liste(movie["overview"])
l_descp = l_descp.dropna()
l_descp.head()
```

On utilise le compteur pour compter le nombre de fois où un mot apparait et l'on transforme cette liste en séries pour pouvoir la classé par ordre décroissant. On trouve que THE est le mot le plus utilisé dans les desciptions.

```python
from collections import Counter
compte = Counter(l_descp.str.split(" ").sum())
pd.Series(compte).sort_values(ascending = False)
```

# Question_7

Je créer une DataFrame movie2 qui est une copie, pour ne pas altérer le fichier original. Je transforme la colone production_countries de manière à "lire" les JSON

```python
import json as js
movie2 = pd.read_csv("tmdb_5000_movies.csv").copy()
movie2["production_countries"] = movie2["production_countries"].apply(js.loads)
```

Avec des boucles imbriqués on arrive à créer une liste des "iso" des pays, ils apparaissent dans la liste autant de fois qu'ils apparaissent dans la colonne. 

```python
list_pays=[]
for index, i in zip(movie2.index, movie2["production_countries"]):
    for j in range(len(i)): 
        list_pays.append((i[j]["iso_3166_1"]))
```

On fait la même chose avec les noms complets des pays

```python
list_name=[]
for index, i in zip(movie2.index, movie2["production_countries"]):
    for j in range(len(i)): 
        list_name.append((i[j]["name"]))
```

A partir de ces deux listes on créer une nouvelle DataFrame que l'on appelle ici porduction

```python
production = pd.DataFrame({
    "Pays": list_pays,
    "Name": list_name,
})
production.head()
```

On affiche avec un diagramme la répartition des pays en fonction du nombre de films produits, on voit clairement que se sont les états-unis qui produisent le plus de films.

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1,1, figsize = (10,50))
production['Pays'].value_counts().plot(kind= "pie")
```

Pour compléter le diagramme, on utilise une nouvelle fois un compteur pour créer une liste que l'on va ensuite trier dans l'ordre décroissant.

```python
l_pays = production['Pays']
compte_pays = Counter(l_pays.str.split(" ").sum())
pd.Series(compte_pays).sort_values(ascending = False)
```
